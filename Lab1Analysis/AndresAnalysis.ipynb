```python
import matplotlib.pylot as plt
import numpy as np
def extractData(filename):
    ''' Extracting captured data from the lab1 npz array from SDR'''
    raw_data = np.load(filename)
    return raw_data['arr_0'][0]
```

```python
CB = ['#377eb8', '#ff7f00', '#4daf4a',
                  '#f781bf', '#a65628', '#984ea3',
                  '#999999', '#e41a1c', '#dede00']
#Colorblind friendly colors (?)
```

```python
path = '../Lab1Data/Section5_2/'

data_400khz_1300khz = extractData(path+'lab_1_data_400khz_1300khz.npz')
data_400khz_2000khz = extractData(path+'lab_1_data_400khz_2000khz.npz')
data_400khz_3100khz = extractData(path+'lab_1_data_400khz_3100khz.npz')
data_1500khz_1300khz = extractData(path+'lab_1_data_1500khz_1300khz.npz')
data_1500khz_2000khz = extractData(path+'lab_1_data_1500khz_2000khz.npz')
data_1500khz_3100khz = extractData(path+'lab_1_data_1500khz_3100khz.npz')
data_1500khz_1300khz_2 = extractData(path+'lab_1_data_1500khz_1300khz_2.npz')
data_1500khz_2000khz_2 = extractData(path+'lab_1_data_1500khz_2000khz_2.npz')
data_1500khz_3100khz_2 = extractData(path+'lab_1_data_1500khz_3100khz_2.npz')
data_1500khz_1300khz_3 = extractData(path+'lab_1_data_1500khz_1300khz_3.npz')
data_1500khz_2000khz_3 = extractData(path+'lab_1_data_1500khz_2000khz_3.npz')
data_1500khz_3100khz_3 = extractData(path+'lab_1_data_1500khz_3100khz_3.npz')
```

```python
plt.plot(data_1500khz_3100khz_2, marker = 'o',  markersize=3, c = CB[0], label = 'Sample Rate 3100KHz')
plt.plot(data_1500khz_2000khz_2, marker = 'o',  markersize=3, c = CB[4], label = 'Sample Rate 2000KHz')
plt.plot(data_1500khz_1300khz_2, marker = 'o',  markersize=3, c = CB[8], label = 'Sample Rate 1300KHz')
plt.title('Measurements of 1500KHz Signal')
plt.xlabel("sample")
plt.ylabel("Maybe Volts")
plt.legend()
#plt.xlim(1870,2047)
plt.show()
```

```python
plt.plot(data_1500khz_3100khz_3, marker = 'o',  markersize=3, c = CB[0], label = 'Sample Rate 3100KHz')
plt.plot(data_1500khz_2000khz_3, marker = 'o',  markersize=3, c = CB[4], label = 'Sample Rate 2000KHz')
plt.plot(data_1500khz_1300khz_3, marker = 'o',  markersize=3, c = CB[8], label = 'Sample Rate 1300KHz')
plt.title('Measurements of 1500KHz Signal')
plt.xlabel("sample")
plt.ylabel("Maybe Volts")
plt.legend()
plt.xlim(1870,2047)
plt.show()
```
